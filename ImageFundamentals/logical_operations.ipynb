## Logical Operations On The Images
Logical operations on images involve performing pixel-wise operations based on logical rules or conditions. These operations are typically used for image processing tasks such as image blending, masking, thresholding, and bitwise operations

Image Thresholding:
   Binary Thresholding: Compares each pixel value to a threshold value. If the pixel value is greater than or equal to the threshold, it is set to the maximum value (e.g., 255) in the resulting binary image; otherwise, it is set to the minimum value (e.g., 0). This operation can be used to create binary masks or segment images based on intensity.


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image in grayscale
image_path = 'multiple_blocks.png'
img = cv2.imread(image_path, 0)

# Apply thresholding
_, binary_image_1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
_, binary_image_2 = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY)

# Create a figure placeholder
fig = plt.figure(figsize=(12, 5))

# show original image on a subplot
fig.add_subplot(131)
plt.title(' image ')
plt.set_cmap('gray')
plt.imshow(img)

# show original image on a subplot
fig.add_subplot(132)
plt.title(' binary image ')
plt.set_cmap('gray')
plt.imshow(binary_image_1)

fig.add_subplot(133)
plt.title(' binary image ')
plt.set_cmap('gray')
plt.imshow(binary_image_2)
```

## Logical Operation

Bitwise Operations:

- Bitwise AND: Performs a bitwise AND operation between the pixel values of two images. This operation is commonly used for extracting specific bit patterns or performing bitwise masking.

- Bitwise OR: Performs a bitwise OR operation between the pixel values of two images. This operation can be used for combining or merging bit patterns.

- Bitwise XOR: Performs a bitwise XOR (exclusive OR) operation between the pixel values of two images. This operation can be used for detecting changes or differences between two images.

```python
# Perform logical operations
bitwise_and = cv2.bitwise_and(binary_image_1, binary_image_2)
bitwise_or = cv2.bitwise_or(binary_image_1, binary_image_2)
bitwise_xor = cv2.bitwise_xor(binary_image_1, binary_image_2)
bitwise_not = cv2.bitwise_not(binary_image_1)
```

```python
# Create a figure placeholder
fig = plt.figure(figsize=(14, 8))

# show original image on a subplot
fig.add_subplot(151)
plt.title(' image ')
plt.set_cmap('gray')
plt.imshow(img)

# show original image on a subplot
fig.add_subplot(152)
plt.title(' bitwise and image ')
plt.set_cmap('gray')
plt.imshow(bitwise_and)

fig.add_subplot(153)
plt.title(' bitwise or image ')
plt.set_cmap('gray')
plt.imshow(bitwise_or)

fig.add_subplot(154)
plt.title(' bitwise xor image ')
plt.set_cmap('gray')
plt.imshow(bitwise_xor)

fig.add_subplot(155)
plt.title(' bitwise not image ')
plt.set_cmap('gray')
plt.imshow(bitwise_not)
```
